```python
import pandas as pd
```

```python
obce=pd.read_csv('OBCE_GPS.csv',delimiter=",",  encoding='utf-8')
```

```python
import random as rn
```

```python
gen_obce=rn.sample(range(len(obce)),25)
```

```python
gen_obce
```

```python
vyber=obce.ix[gen_obce]
vyber[:10:]
```

```python
import folium
```

```python
stred=[obce[obce['OBEC']=='Zvolen']['LAT'].iloc[0],obce[obce['OBEC']=='Zvolen']['LNG'].iloc[0]]
```

```python
hash1={}
```

```python
mapa = folium.Map(location=stred, zoom_start=7)
cluster = folium.MarkerCluster().add_to(mapa)
folium.Marker(stred, popup=obce[obce['OBEC']=='Zvolen']['DOBEC'].iloc[0]).add_to(cluster)
for poradie,data in vyber.iterrows():
    folium.Marker([data["LAT"], data["LNG"]], popup=data["DOBEC"]).add_to(cluster)
```
